```python
import pandas as pd 

data = pd.read_csv("C:/Users/charl/OneDrive - Worcester Polytechnic Institute (wpi.edu)/Desktop/fuel_emissions1csv.csv", low_memory=False)
dfClean = pd.DataFrame(data)
yearly_mean_methane = dfClean.groupby("Reporting Year")["Fuel Methane (CH4) emissions (mt CO2e)"].mean()
yearly_mean_n2o = dfClean.groupby("Reporting Year")["Fuel Nitrous Oxide (N2O) emissions (mt CO2e)"].mean()


dfClean1 = dfClean.copy()
dfClean1["Yearly Methane Avg"] = dfClean["Reporting Year"].map(yearly_mean_methane) #add yearly means for methane and no2 to data
dfClean1["Yearly NO2 Avg"] = dfClean["Reporting Year"].map(yearly_mean_n2o)
dfClean1.to_csv("fuel_cleaned.csv", index=False)


```

# Descriptive Statistics 

```python
avg_methane = dfClean1["Fuel Methane (CH4) emissions (mt CO2e)"].mean()
avg_n2o = dfClean1["Fuel Nitrous Oxide (N2O) emissions (mt CO2e)"].mean()

med_methane = dfClean1["Fuel Methane (CH4) emissions (mt CO2e)"].median()
med_n2o = dfClean1["Fuel Nitrous Oxide (N2O) emissions (mt CO2e)"].median()

industry_mode = dfClean["Industry Type (sectors)"].mode()
fuel_mode = dfClean["General Fuel Type"].mode()
state_mode = dfClean["State"].mode()


#create copy of dfClean1 to drop the years
dfClean2 = dfClean1.copy()
dfClean2.drop("Reporting Year", axis=1, inplace=True)
stdev = dfClean2.std()
var = dfClean2.var()

print(avg_methane)
print(avg_n2o)
print(med_methane)
print(med_n2o)

print(industry_mode) #power plants
print(fuel_mode) #natural gas
print(state_mode) #Texas 

print(stdev)
print(var)
```

# Below is max scaled (normalized data) idk if we want to normalize it 


```python
df_reportYr = pd.DataFrame(dfClean["Reporting Year"]) #make year its own dataframe so the year will not be normalized

dfClean_maxScaled = dfClean1.copy()
dfClean_maxScaled.drop(["State", "General Fuel Type", "Industry Type (sectors)", "Reporting Year"], 
                       axis=1, inplace=True) #drop year and non numeric columns

dfClean_maxScaled = dfClean_maxScaled.apply(pd.to_numeric, errors='coerce') #convert the remaining columns to numeric, 
                                                                            #and et errors='coerce' to handle non-numeric values by converting them to NaN.

for column in dfClean_maxScaled.columns:
    dfClean_maxScaled[column] = dfClean_maxScaled[column]  / dfClean_maxScaled[column].abs().max() #normalize data

dfClean_maxScaled = pd.concat([df_reportYr, dfClean_maxScaled], axis=1) #add reporting year back in 

      
# view normalized data
display(dfClean_maxScaled)

dfClean_maxScaled.to_csv("fuel_maxScaledcsv.csv", index=False)
```

# min-max normalization

```python
dfClean_minMax = dfClean1.copy()

dfClean_minMax.drop(["State", "General Fuel Type", "Industry Type (sectors)", "Reporting Year"], axis=1, inplace=True)
dfClean_minMax = dfClean_minMax.apply(pd.to_numeric, errors='coerce')

for column in dfClean_minMax.columns:
    dfClean_minMax[column] = (dfClean_minMax[column] - dfClean_minMax[column].min()) / (dfClean_minMax[column].max() - dfClean_minMax[column].min())   
    

dfClean_minMax = pd.concat([df_reportYr, dfClean_minMax], axis=1) #add reporting year back in 

  
# view normalized data
print(dfClean_minMax)

dfClean_minMax.to_csv("fuel_minMaxcsv.csv", index=False)
```

# Z score

```python
dfClean_z = dfClean1.copy()

dfClean_z.drop(["State", "General Fuel Type", "Industry Type (sectors)", "Reporting Year"], axis=1, inplace=True)
dfClean_z = dfClean_z.apply(pd.to_numeric, errors='coerce')

for column in dfClean_z.columns:
    dfClean_z[column] = (dfClean_z[column] - dfClean_z[column].mean()) / dfClean_z[column].std()    
    
dfClean_z = pd.concat([df_reportYr, dfClean_z], axis=1) #add reporting year back in 
  
# view normalized data   
display(dfClean_z)

#csv file 
dfClean_z.to_csv('fuelClean_z.csv', index=False)
```
